# 1. O que é Pokémon?

Os Pokémon são criaturas fictícias que habitam um mundo fictício chamado de Mundo Pokémon. Eles são os principais protagonistas da franquia de mídia Pokémon, que inclui jogos de vídeo, desenhos animados, filmes, card games e uma variedade de outros produtos.

Cada Pokémon é caracterizado por suas próprias individualidades, incluindo habilidades, status, egg groups e tipos:

- Habilidades: As habilidades dos Pokémon são características distintas que podem afetar o curso das batalhas ou interações no jogo. Algumas habilidades aumentam os atributos do Pokémon, outras alteram os efeitos dos movimentos ou têm efeitos passivos.
- Status: Os status dos Pokémon referem-se aos seus atributos básicos, como HP (pontos de vida), ataque, defesa, sp_attack (ataque especial), sp_defense (defesa especial) e velocidade. Cada Pokémon tem seus próprios valores de status, que podem variar amplamente entre as espécies.
- Egg Groups: Os egg groups definem com quais outros Pokémon um determinado Pokémon pode acasalar para produzir ovos. Esses grupos são usados para determinar a compatibilidade de reprodução entre Pokémon no jogo.
- Tipos: Os Pokémon têm um ou dois tipos, que determinam suas fraquezas e resistências a diferentes tipos de ataques. Existem 18 tipos diferentes no total, incluindo tipos como grama, fogo, água, elétrico, psíquico, dragão, entre outros. A combinação de tipos de um Pokémon pode afetar significativamente sua eficácia em batalha. Exemplo: Um Pokémon do tipo grama recebe dano dobrado de um ataque tipo fogo, mas recebe metade do dano de um pokémon tipo água. A combinação de dois tipos pode trazer imunidades diferentes e resistências de até 1/4 de dano, mas também receber 4 vezes o dano de determinados tipos.
- Características físicas: Altura e peso também podem influenciar nas batalhas, pois existem golpes e habilidades que levam em consideração essas características
- Porcentagem na natureza de machos e fêmeas: Alguns Pokémon têm diferenças de gênero em suas proporções de machos e fêmeas na natureza. Enquanto alguns Pokémon têm uma divisão de gênero de 50/50, outros podem ter uma proporção maior de machos ou fêmeas. Isso pode afetar a criação e a obtenção de Pokémon com determinados atributos ou movimentos, além da eficácia de movimentos que usam esse atributo com base.
- Pokémon sem gênero: Alguns Pokémon são considerados sem gênero e não têm distinção entre macho e fêmea. Isso geralmente se aplica a Pokémon lendários, míticos e alguns Pokémon semelhantes a objetos.

Essas características únicas contribuem para a diversidade e complexidade do mundo Pokémon, tornando cada espécie única e oferecendo uma ampla gama de estratégias e possibilidades no jogo.

Além disso existem características (também únicas a cada espécie ou até indivíduo) que dão mais informações sobre o contexto em que ele foi inserido

- Gerações: As gerações referem-se aos diferentes períodos de lançamento dos jogos principais da série Pokémon. Cada geração, geralmente, coincide com o lançamento de novos jogos principais da Nintendo. Cada geração introduz novos Pokémon, mecânicas de jogabilidade, regiões exploráveis e histórias. Atualmente a franquia possui 9 gerações.
- Lendário/Sub Lendário/Mítico: Esses termos são usados para descrever certos Pokémon que são considerados especiais ou únicos dentro do universo Pokémon. Os Pokémon lendários são poderosos e geralmente têm um papel significativo na narrativa do jogo. Os sub lendários são uma categoria menos restrita, mas ainda incluem Pokémon raros e poderosos, com participações em histórias secundárias. Os Pokémon míticos são extremamente raros e associados à lendas e mitos no universo, e geralmente só podem ser obtidos através de eventos especiais ou distribuição online.
- Catch Rate: A taxa de captura é uma medida da facilidade com que um jogador pode capturar um Pokémon selvagem durante uma batalha. Quanto mais alta a taxa de captura, mais fácil é capturar o Pokémon. A taxa de captura varia de acordo com o Pokémon e pode ser influenciada por fatores como o status de saúde do Pokémon selvagem, o uso de Poké Balls específicas e outras condições.
- Base Experience: A experiência base é a quantidade de experiência que um jogador recebe ao derrotar um Pokémon selvagem. Cada Pokémon tem uma quantidade específica de experiência base associada a ele, que é usada para calcular a quantidade de experiência que os Pokémon do jogador recebem após uma batalha.
- Egg Cycles: Os ciclos de ovos representam o tempo necessário para que um ovo de Pokémon choque e produza um Pokémon recém-nascido. Cada espécie de Pokémon tem um número específico de ciclos de ovos necessários para chocar, que pode variar de algumas centenas a milhares de ciclos, dependendo da espécie. Um ciclo possui 255-257 passos (dependendo do jogo)

- ability_hidden: nome da habilidade escondida, se ele tiver (texto)

- hp: o HP base do pokémon (inteiro)
- attack: o ataque base do pokémon (inteiro)
- defense: a defesa base do pokémon (inteiro)
- sp_attack: o ataque especial base do pokémon (inteiro)
- sp_defense: a defesa especial base do pokémon (inteiro)
- speed: a velocidade base do pokémon (inteiro)
- catch_rate: o catch rate do pokémon (inteiro)
- base_friendship: a amizade base do pokémon quando é capturado (inteiro)
- base_experience: a experiência base do pokémon quando é capturado (inteiro)
- growth_rate: o tipo da curva de crescimento de experiência (texto)
- egg_type_number: números de grupos de tipo de ovo que o pokémon possui (inteiro: 0, 1 ou 2)
- egg_type_1: nome do primeiro egg type que o pokémon possui (texto)
- egg_type_2: nome do segundo egg type que o pokémon pode ter (texto)
- percentage_male: o percentual de indivíduos macho daquela espécie na natureza, vazio se o pokémon não tiver gênero (texto)
- egg_cycles: número de cilos (255-257 passos) necessários para rachar o ovo do pokémon (inteiro)

# 5. Visualização de dados

1 - Qual o tipo de pokémon que mais aparece?

Para responder essa pergunta não é possível apenas contar os tipos, pois existem duas colunas (type_1 e type_2). Para isso, foi necessário primeiro realizar a concatenação das colunas de tipos e em seguida realizar a contagem usando o value_counts

```python
# concatenando os tipos
tipos_concatenados = pd.concat([df_pokemon['type_1'], df_pokemon['type_2']])
# contando os tipos
contagem_tipos = tipos_concatenados.value_counts()
# mostrando a contagem de quantos pokémon existem para cada tipo
print(contagem_tipos)
```

Com isso conseguimos ver que o tipo que mais aparece é o tipo Water (Água) com 154 no total

2 - Qual a habilidade que mais aparece?

Para responder essa pergunta, também não é possível apenas contar as habilidades, pois existem três colunas (ability_1, ability_2 e	ability_hidden). Para isso, foi necessário primeiro realizar a concatenação das colunas e em seguida realizar a contagem usando o value_counts

```python
# juntando as colunas usando o melt
habilidades = df_pokemon.melt(value_vars=['ability_1', 'ability_2', 'ability_hidden'])['value']
# contando e juntando
contagem_habilidades = habilidades.value_counts()
# mostrando as 5 que mais aparecem
contagem_habilidades.head(5)
```

Após a análise, conseguimos encontrar que Swift Swim é a habilidade mais encontrada entre os pokémon, com o total de 43 aparições.

3 - Qual é mediana do poder de cada um dos tipos primários dos pokemons?

```python
# criano nova coluna com os tipos combinados e separados por _
df_pokemon['combined_types'] = df_pokemon['type_1'] + '_' + df_pokemon['type_2']

# se algum tipo for NaN, preencha com o valor da primeira coluna
df_pokemon['combined_types'].fillna(df_pokemon['type_1'], inplace=True)
# checando se de fato está criado e preenchido corretamente
print(df_pokemon.head(10))
```

```python
# criando uma lista com todos os tipos
lista_tipos = ['Normal', 'Fighting', 'Flying', 'Poison', 'Ground', 'Rock', 'Bug', 'Ghost', 'Steel', 'Fire', 'Water', 'Grass', 'Electric', 'Psychic', 'Ice', 'Dragon', 'Dark', 'Fairy']
#criando um novo dataframe com mediana de cada tipo
mediana_tipos = pd.DataFrame(columns=['tipo', 'mediana'])
# vendo se de fato está criado
mediana_tipos.head(5)
```

```python
# a ideia é procurar cada tipo na lista de tipos, filtrar o dataframe pra aquele
# tipo e fazer a mediana
for tipo in lista_tipos:
    # filtrando o dataframe onde o tipo está presente na coluna combined_types
    dataframe_filtrado = df_pokemon[df_pokemon['combined_types'].str.contains(tipo, na=False)]

    # calculando a mediana da coluna total_points para o tipo específico
    mediana = dataframe_filtrado['total_points'].median()
    # criando o conjunto que vai ser adicionado no dataframe de medianas
    conjunto = {'tipo': tipo, 'mediana': mediana}

    # adicionando o conjunto de tipo e mediana ao dataframe de medianas
    mediana_tipos = pd.concat([mediana_tipos, pd.DataFrame([conjunto])], ignore_index=True)

mediana_tipos = mediana_tipos.sort_values(by='mediana',ascending=False)
print(mediana_tipos)
```

```python
# Criando um df que recebe um agrupamento dos tipos pela mediana dos total_points
media_tipos = df_pokemon.groupby('type_1')['total_points'].agg(mean=np.mean)

# Ordenando o DF das maiores medianas para as menores
media_tipos.sort_values(by='mean',ascending=False)
```

4 - Quais são os 20 pokemons com maior HP?

```python
# 20 Pokémon com os maiores valores na coluna hp
top_hp = df_pokemon.nlargest(20, 'hp')

# lista com todos os tipos
lista_tipos = ['Normal', 'Fighting', 'Flying', 'Poison', 'Ground', 'Rock', 'Bug', 'Ghost', 'Steel', 'Fire', 'Water', 'Grass', 'Electric', 'Psychic', 'Ice', 'Dragon', 'Dark', 'Fairy']

# criando um dicionário para contar a quantidade de cada tipo
contador_tipos= {type_: 0 for type_ in lista_tipos}

# contando a quantidade de cada tipo no dataframe
for type_ in lista_tipos:
    contador_tipos[type_] = (top_hp['type_1'] == type_).sum() + (top_hp['type_2'] == type_).sum()

# montando um df a partir do dicionário contador
contador_tipos_df = pd.DataFrame(list(contador_tipos.items()), columns=['Tipo', 'Quantidade'])
contador_tipos_df = contador_tipos_df.sort_values(by='Quantidade',ascending=False)

print(contador_tipos_df)
```

Com isso, encontramos que dentre os 20 pokémon com maior HP, 7 deles são do tipo Dragão, sendo esse o tipo que mais aparece.
